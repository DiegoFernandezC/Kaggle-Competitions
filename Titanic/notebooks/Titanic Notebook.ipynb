```python
import pandas as pd
```

```python
data_train = pd.read_csv("../datasets/train.csv") 
```

```python
data_train.head()
```

### Transformamos variable categorica y eliminamos columna redundante
* 0 -> Female
* 1 -> Male

```python
dummy_gender = pd.get_dummies(data_train["Sex"], prefix="Sex").iloc[:, 1:]

dummy_gender.head()
```

### join con el dataset y eliminamos la columna Sex

```python
data_train = data_train.drop(["Sex"], axis=1)

data_train.head()
```

```python
data_train = data_train.join(dummy_gender)

data_train.head()
```

### Tratamiento variable categorica Embarked

* S queda determinado por C y Q

```python
dummy_embarked = pd.get_dummies(data_train["Embarked"], prefix="Embarked")

dummy_embarked.head()
```

```python
dummy_embarked = dummy_embarked.iloc[:, :2]

dummy_embarked.head()
```

```python
data_train = data_train.drop(["Embarked"], axis=1)

data_train.head()
```

```python
data_train = data_train.join(dummy_embarked)

data_train.head()
```
